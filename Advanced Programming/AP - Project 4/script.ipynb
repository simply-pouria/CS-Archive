```python
class MyDataFrame:
    def __init__(self, data, columns):
        self.df =pd.DataFrame(data, columns=columns)
        
    def add_row(self, row_dic):
        self.df.loc[len(self.df.index)] = row_dic
        
    def remove_row(self, row_dic):
        self.df.drop(row_dic, axis = 0, inplace=True)
        
    def filter_by_column(self, column_name, value):
        return self.df[self.df[column_name] == value]
    
    def to_dict(self):
        return self.df.to_dict(orient='records')
    
    def save_csv(self, file_path):
        self.df.to_csv(file_path, index=False)
        
# this method is also possible (instead of inputting dictionaries) when we are also inputting columns separately
df = MyDataFrame([[1,2,3], [4,5,6], [7,8,9], ['a','b','c']],["first", "second", "third"])

        
        
        
```

# Exercise 4

```python
class MyDataFrame:
    def __init__(self, data):
        
        if type(data) == str:
            self.from_csv(file_path=data)
        elif type(data) == dict:
            self.from_dict(data)
        
    def add_row(self, row_dic):
        self.df.loc[len(self.df.index)] = row_dic
        
    def remove_row(self, row_dic):
        self.df.drop(row_dic, axis = 0, inplace=True)
        
    def filter_by_column(self, column_name, value):
        return self.df[self.df[column_name] == value]
    
    def to_dict(self):
        return self.df.to_dict(orient='records')
    
    def save_csv(self, file_path):
        self.df.to_csv(file_path, index=False)
        
    def from_dict(self, dic):
        self.df = pd.DataFrame(dic)
        
    def from_csv(self, file_path):
        self.df = pd.read_csv(file_path)
        
```
